```python
import requests
from bs4 import BeautifulSoup

# URL of the website to scrape
url = "https://www.visitmyphilippines.com/index_title_UsefulLinks_func_all_pid_13_tbl_1.html"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the specific section containing useful links (adjust selectors based on the structure of the website)
    links_section = soup.find_all('a')  # Finds all anchor tags
    
    # Extract and display links and their text
    useful_links = []
    for link in links_section:
        href = link.get('href')  # Get the hyperlink
        text = link.get_text(strip=True)  # Get the text
        if href:  # Ensure the link exists
            useful_links.append({"text": text, "url": href})
    
    # Print the extracted useful links
    for item in useful_links:
        print(f"Text: {item['text']}, URL: {item['url']}")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the website to scrape
url = "https://www.visitmyphilippines.com/index_title_VisitMyPhilippines_func_all_Page_1.html/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the specific section containing useful links (adjust selectors based on the structure of the website)
    links_section = soup.find_all('a')  # Finds all anchor tags
    
    # Extract and store links in a list
    useful_links = []
    for link in links_section:
        href = link.get('href')  # Get the hyperlink
        text = link.get_text(strip=True)  # Get the text
        if href:  # Ensure the link exists
            useful_links.append({"Text": text, "URL": href})
    
    # Create a DataFrame from the useful links
    df = pd.DataFrame(useful_links)
    
    # Display the DataFrame
    print(df)
    
    # Save the DataFrame to a CSV file (optional)
   
    print("Data saved to 'useful_links.csv'")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

df
```
